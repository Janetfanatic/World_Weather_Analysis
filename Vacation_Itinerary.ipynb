```python
import gmaps.datasets
gmaps.configure(api_key=g_key)

#Latitude-Longitude pairs
tivoli = (41.96, 12.8)
vieste = (41.88, 16.18)
crotone = (39.09, 17.12)
marsala = (37.8, 12.44)

fig = gmaps.figure()
tivoli2marsala_via_vieste_and_crotone = gmaps.directions_layer(
    tivoli, marsala, waypoints=[vieste, crotone],
    travel_mode='DRIVING')
fig.add_layer(tivoli2marsala_via_vieste_and_crotone)
fig
```

```python
weatherpy_travel_map_markers = pd.concat([city_one, city_two, city_three, city_four])
weatherpy_travel_map_markers
```

```python
weatherpy_travel_map_markers.drop_duplicates(inplace = True)
weatherpy_travel_map_markers
```

```python
# Create the output file (CSV).
output_data_file = "weather_data/WeatherPy_itinerary.csv"
# Export the City_Data into a CSV.
weatherpy_travel_map_markers.to_csv(output_data_file)
```

```python
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Description</dt><dd>{Current Description} and {Max Temp} °F</dd>
</dl>
"""
# Store the DataFrame Row.
weatherpy_travel_info = [info_box_template.format(**row) for index, row in weatherpy_travel_map_markers.iterrows()]
```

```python
# Add a heatmap of temperature for the vacation spots and a pop-up marker for each city.
# Configure gmaps to use your Google API key.

locations = weatherpy_travel_map_markers[["Lat", "Lng"]]
max_temp = weatherpy_travel_map_markers["Max Temp"]
fig = gmaps.figure(center=(30.0, 31.0), zoom_level=1.5)
heat_layer = gmaps.heatmap_layer(locations, weights=max_temp,dissipating=False,
             max_intensity=300, point_radius=4)
marker_layer = gmaps.marker_layer(locations, info_box_content=weatherpy_travel_info)
fig.add_layer(heat_layer)
fig.add_layer(marker_layer)

# Call the figure to plot the data.
fig
```
